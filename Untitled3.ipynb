```python
import pandas as pd
```

```python
df = pd.read_csv("Datasets/students_adaptability_level_online_education.csv")
```

```python
df['Network Type'].unique()
```

```python
df['Network Type'] = df['Network Type'].map({"4G":2, "3G":1, "2G": 0})
```

```python
df['Gender'] = df['Gender'].map({'Girl':0, "Boy": 1})
```

```python
sns.violinplot(df,y = 'Gender')
```

```python
sns.boxplot(df, y = 'Gender')
```

```python
def encode(data):
    encoding = []
    for feature in data:
        unique = data[feature].unique()
        unique.sort()
        maping = {value: key for key, value in enumerate(unique)}
        data[feature] = data[feature].map(maping)
        encoding.append([feature,maping])
    return (data, encoding)
        
```

```python
df_encoded, encodings = encode(df.astype(dtype=object))
```

```python
df_encoded
```

```python
encodings
```

```python
sns.violinplot(df_encoded, y = 'Adaptivity Level', hue = 'IT Student')
```

```python
sns.violinplot(df_encoded, y = 'Adaptivity Level', hue = 'Internet Type')
```

```python
sns.violinplot(df_encoded, y = 'Adaptivity Level', hue = 'Device')
```

```python
sns.pairplot(df_encoded)
```

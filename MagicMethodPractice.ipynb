# 1) Smart Shopping Cart (Operator Overloading)
- A shopping cart should behave like a real object.
- You must allow adding items with +, removing using -, and printing in a readable way.
    - Requirements
        - Class Cart holds items (name, price, qty)
        - cart + item → adds item
        - cart - item → removes item (if exists)
        - str(cart) shows total items & bill

```python
class cart:
    def __init__(self):
        self.items = []
        self.total_bill = 0
    def __add__(self,others):
        self.items.append(others)
        self.total_bill += others[1]*others[2]
    def __sub__(self,others):
        if others < len(self.items):
            removed = self.items.pop(others-1)
            self.total_bill -= removed[1]*removed[2]
        else:
            print("Item does not exists....!!!")
    def __str__(self):
        for index, item in enumerate(self.items):
            print(F"{index+1}. {item[0]} || {item[1]} || {item[2]}")
        return f"Total Items: {len(self.items)} || Total Price: {self.total_bill}"
```

```python
my_cart = cart()
```

# Attendance Sytem (Use len() and Membership in)
- attendance+'Ali' -> Mark Attendance
- len(attendance) -> Total attendance
- 'Ali' in attendance -> True or False
## Note the class attendance stores student name internally

```python
class attendance:
    def __init__(self):
        self.students = []
    def __add__(self,others):
        self.students.append(others.lower())
        print(f"Student: {others} is marked as present...")
    def __len__(self):
        return len(self.students)
    def __contains__(self,others):
        return others.lower() in self.students
    def __str__(self):
        for std in self.students:
            print(f"Student name: {std}")
        return f"{len(self.students)} students are marked as present for today.."
    
    
```

```python
today = attendance()

```

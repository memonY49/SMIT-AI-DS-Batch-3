# Decorators

```python
def greet():
    def desplay():
        print("This is the class of SMIT")
        
    print('Welcome')
    desplay()
    print('GoodBye')
```

```python
greet()
```

```python
def decorate_greet(func):
    def wrapper(*args,**kwargs):
        print("welcome")
        func(*args,**kwargs)
        print('Goodbye')
    return wrapper

    
def confirmation(func):
    def wrapper(*args,**kwargs):
        userconfirm = input("Enter y to confirm and n to cancel: ")
        if userconfirm.lower() == 'y':
            func(*args,**kwargs)
    return wrapper
```

```python
@confirmation
@decorate_greet
def desplay():
        print("This is the class of SMIT")
```

```python
desplay()
```

```python
@decorate_greet
def display_name(name = 'SMIT'):
    print(name)
```

```python
display_name(name = 'Yasir')
```

```python
def thankyou(func):
    def wrapper(*args,**kwargs):
        myFunc= func(*args,**kwargs)
        print("Thanks for using this ATM...")
        return myFunc
    return wrapper


def confirm(func):
    def wrapper(*args,**kwargs):
        userinput = input(f"Please confirm before {func.__name__} transaction (y/n): ")
        if userinput.lower() == 'y':
            return func(*args,**kwargs)
        else:
            print("User transaction is canceled...")
            return None
    return wrapper

Accounttitle = 'Yasir Nawaz'
Balance = 10000

@thankyou
def accountTitle(title):
    print(title)

@thankyou
@confirm
def check_balance(balance):
    print(f"Your current balance is: {balance}")

@thankyou
@confirm
def deposit(balance):
    amount = int(input("Enter your amount to deposit: "))
    balance += amount
    print(f"hello {balance}")
    return balance

@thankyou
@confirm
def withdraw(balance):
    amount = int(input("Enter your amount to withdraw: "))
    balance -= amount
    return balance
```

```python
class BankAccount:
    def __init__(self, Title, amount = 0):
        self.Title = Title
        self.Amount = amount
        
    def thankyou(func):
        def wrapper(*args,**kwargs):
            myFunc= func(*args,**kwargs)
            print("Thanks for using this ATM...")
            return myFunc
        return wrapper


    def confirm(func):
        def wrapper(*args,**kwargs):
            userinput = input(f"Please confirm before {func.__name__} transaction (y/n): ")
            if userinput.lower() == 'y':
                return func(*args,**kwargs)
            else:
                print("User transaction is canceled...")
                return None
        return wrapper

    @thankyou
    def accountTitle(self):
        print(self.Title)
    
    @thankyou
    @confirm
    def check_balance(self):
        print(f"Your current balance is: {self.Amount}")
    
    @thankyou
    @confirm
    def deposit(self):
        amount = int(input("Enter your amount to deposit: "))
        self.Amount += amount
    
    @thankyou
    @confirm
    def withdraw(balance):
        amount = int(input("Enter your amount to withdraw: "))
        self.Amount -= amount
        
```

```python
Yasir = BankAccount("Yasir")
Ahmed = BankAccount("Ahmed",amount=30000)
```

```python
Yasir.check_balance()
```

```python
Ahmed.deposit()
```

```python
for path, listdir, listfile in  os.walk('database'):
    print("Path:",path)
    print(f"Total dir: {len(listdir)} : {','.join(listdir)}")
    textfile = []
    for file in listfile:
        if os.path.splitext(file)[1] == '.txt':
            textfile.append(file)
    print(f"Total file: {len(textfile)} : {','.join(textfile)}")
    print("*"*10)
```

```python
with open('employee.txt','r') as file:
    for emp in file.readlines():
        emp_data = emp.strip().split(',')
        diprt_dir = "database\\"+emp_data[2]
        user_dir = f"database\\{emp_data[2]}\\{emp_data[1]}"
        
        if not os.path.exists(diprt_dir):
            os.mkdir(diprt_dir)
        if not os.path.exists(user_dir):
            os.mkdir(user_dir)
        
```

```python
orders['order_date'] = pd.to_datetime(orders.order_date)
```

```python
orders['year'] = orders.order_date.dt.year
```

```python
orders
```

```python
orders['months'] = orders.order_date.dt.month
```

```python
order_group = orders.groupby('year')
```

```python
orders
```

```python
m_2022_groupe = order_group.get_group(2022).groupby('months')
m_2023_groupe = order_group.get_group(2023).groupby('months')
m_2024_groupe = order_group.get_group(2024).groupby('months')
```

```python
for year in orders.year.unique():
    year_grp = order_group.get_group(year)
    print(f"Year {year}: Total Quantity Soled: {year_grp.quantity.sum()}")
    counts = year_grp.customer_id.value_counts()
    for index ,value in zip(counts.index,counts.values):
        if value >= 5:
            print(f'{index}: {value}')
    print('*'*10)
```

```python
customers = pd.read_csv('customers.csv')
```

```python
customers
```

```python
orders_merge = orders.merge(customers, on = 'customer_id', how = 'left')
```

```python
for year in orders.year.unique():
    year_grp = order_group.get_group(year)
    print(f"Year {year}: Total Quantity Soled: {year_grp.quantity.sum()}")
    counts = year_grp.customer_id.value_counts()
    for index ,value in zip(counts.index,counts.values):
        if value >= 5:
            name = orders_merge[orders_merge['customer_id'] == index]['name'].iloc[0]
            print(f'{index}: {name}: {value}')
    print('*'*10)
```

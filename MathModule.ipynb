## r: radius of a circle

```python
def areaofcircle(r):
    return math.pi * (math.pow(r,2))
```

```python
def distanceoftwopoints(p1,p2):
    return math.sqrt(math.pow(p1[0]-p2[0], 2) + math.pow(p1[1]-p2[1], 2))
    
```

```python
x1 = 2
y1 = 3
x2 = 4
y2 = 2
print(distanceoftwopoints((x1,y1),(x2,y2)))
```

# create a function that finds is the number a perfect sqrt or not 

```python
def isperfectsqrt(num):
    num_sqrt = math.sqrt(num)
    if num_sqrt == math.floor(num_sqrt):
        return True
    else:
        return False
    
```

```python
isperfectsqrt(25)
```

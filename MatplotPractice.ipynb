```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
x = np.array([1,3,5,7,9,11])
y = np.array([20,30,40,55,60,70])
```

```python
my_plt = plt.plot(x,y, marker = 'x',c='#dd42f5', linestyle = '-', linewidth = 2)
#linestyles  '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
# plt.plot([5,6,7,8,9],[44,55,66,77,88])
plt.xlabel("x-label",labelpad=10)
plt.ylabel("y-label",labelpad=10)
plt.title("My first line chart!!",pad=10,y=1)

# for px,py in zip(x,y):
#     plt.text(x=px,y=py,s=f'p{px}')
    
plt.show()
```

```python
y1 =[]
y2 =[]
x = range(-100,100,10)
for i in x: 
    y1.append(i**2) 
for i in x: y2.append(-i**2) 
```

```python
plt.plot(x,y1,label = 'y1')
plt.plot(x,y2,label = 'y2')
plt.legend()
plt.axvline(x=0, c='r')
plt.axhline(y=0,c='r')
plt.xlim(-200,200)
plt.ylim(-20000,20000)
plt.savefig('multilinechart.png')
plt.show()
```

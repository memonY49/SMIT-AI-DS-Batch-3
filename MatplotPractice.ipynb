```python
y1 =[]
y2 =[]
x = range(-100,100,10)
for i in x: 
    y1.append(i**2) 
for i in x: y2.append(-i**2) 
```

```python
plt.plot(x,y1,label = 'y1')
plt.plot(x,y2,label = 'y2')
plt.legend()
plt.axvline(x=0, c='r')
plt.axhline(y=0,c='r')
plt.xlim(-200,200)
plt.ylim(-20000,20000)
plt.savefig('multilinechart.png')
plt.show()
```

```python
activities = ['eat', 'sleep', 'work', 'play'] 
slices = [8, 4, 9, 5] 
max_index = np.array(slices).argmax()
max = [0.2 if i==max_index else 0 for i in range(0,len(slices))]

max = []
for i in range(0,len(slices)):
    if i == max_index:
        max.append(0.3)
    else:
        max.append(0)


plt.pie(x = slices, labels=activities, autopct = '%1.1f%%', colors=['r','b','g','y'], startangle=120,explode=max)
```

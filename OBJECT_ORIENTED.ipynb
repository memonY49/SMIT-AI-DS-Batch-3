# The Cunstructor Method

```python
class calculator:
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def add(self):
        print(self.a+self.b)
    def sub(self):
        print(self.a-self.b)
    def mul(self):
        print(self.a*self.b)
    def devision(self):
        print(self.a/self.b)
```

```python
cal = calculator(3,7)
```

```python
cal.mul()
```

```python
# generate a list of students where the student datatype should contain its name, studentID, class, dob
# create a method for student_search_by_id, add_multi_students
class human:
    def __init__(self):
        self.hands = 0
        self.fingers = 5
        self.gender = ''
    def speak(self):
        print('is speaking.....')



class student(human):
    def __init__(self,name,id,dob):
        self.name = name
        self.id = id
        self.dob = dob

class listofstudents:
    def __init__(self):
        self.allstudents = []
        
    def addstudent(self,std):
        self.allstudents.append(std)
        
    def printallstudents(self):
        for std in self.allstudents:
            print(f'Name: {std.name}, stdid: {std.id}, dob: {std.dob}')
            
    def searchStdByID(self,std_id):
        for index, std in enumerate(self.allstudents):
            if std_id == std.id:
                return (index,std)
                
    def add_multi_students(self,liststd):
        for std in liststd:
            self.allstudents.append(std)
            
    def updatebyid(self,std_id,name,dob):
        # std = self.searchStdByID(std_id)
        # std[1].name = name
        # std[1].dob = dob
        # self.allstudents[std[0]] = std[1]
        for index, std in enumerate(self.allstudents):
            if std.id == std_id:
                std.name = name
                std.dob = dob
                self.allstudents[index] = std
        
    
```

```python
listofstd = listofstudents()
```

```python
std1 = student('yasir','1','2/10/1997')
std2 = student('yasi1','2','2/10/1997')
std3 = student('yasir2','3','2/10/1997')
std4 = student('yasir3','4','2/10/1997')

listofstd.add_multi_students([std1,std2,std3,std4])
listofstd.addstudent(student('ahmed','5','3/09/1998'))
```

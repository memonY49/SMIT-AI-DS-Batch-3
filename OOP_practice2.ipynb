# How to create a class 

```python
class calculator:
    def __init__(self,val1,val2):
        self.val1 = val1
        self.val2 = val2
    def add(self):
        return self.val1+self.val2
    def sub(self):
        return self.val1-self.val2
    def mul(self):
        return self.val1*self.val2
    def div(self):
        return self.val1/self.val2
        
```

# The constructor
## constructor is a method with same name as a class

```python
cal = calculator(5,6)
```

```python
print(cal.mul())
```

```python
class book:
    def __init__(self,title,auther,publisher,isbn,no_chapt,no_pages):
        self.title = title
        self.auther = auther
        self.publisher = publisher
        self.isbn = isbn
        self.no_chapt = no_chapt
        self.no_pages = no_pages
    def __repr__(self):
        return self.__str__()
        
    def __str__(self):
        details = f"Title: {self.title}\nAuther: {self.auther}\nISBN: {self.isbn}"
        return details
    def __add__(self,other):
        return f"Book1 : {self.title}, Book2: {other.title}"
    def __eq__(self,other):
        if self.title == other.title and self.auther == other.auther and \
        self.publisher == other.publisher and self.isbn == other.isbn and self.no_chapt == other.no_chapt and self.no_pages == other.no_pages:
            return True
        else:
            return False
        

        
    def gettitle(self):
        return self.title
    def settitle(self,new_title):
        self.title = new_title
    def getauther(self):
        return self.auther
    def setauther(self,new_auther):
        self.auther = new_auther
    
        
```

# A Car showroom stores a car details. when the manager prints any car object it should show brand of the car, model, year and color

```python
class student(human):
    def __init__(self,id, name, age,marks = []):
        self.id = id
        self.name = name
        self.age = age
        self.marks = marks


class university:
    def __init__(self, Name, Location, Student = []):
        self.Name = Name
        self.Location = Location
        self.Students = Student
    def add_std(self,std):
        self.Students.append(std)
    def __len__(self):
        return len(self.Students)
    


```

```python
SMIT = university("SMIT University","Hyderabad")
```

```python
SMIT.add_std(student('std_01','Hafiullah',20))
SMIT.add_std(student('std_02','Ahmed',21))
SMIT.add_std(student('std_03','Ali',22))
```

# ATM SYSTEM for add amount in my account I need to use object + (amount) and the amount should be added to the balance property

```python
class account:
    def __init__(self,account_holder, account_no, amount = 0):
        self.account_holder = account_holder
        self.amount = amount
        self.account_no = account_no
    def __add__(self,other):
        self.amount += other
    def __sub__(self,other):
        self.amount -= other
    
```

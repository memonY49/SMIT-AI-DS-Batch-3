# How to create a class 

```python
class calculator:
    def __init__(self,val1,val2):
        self.val1 = val1
        self.val2 = val2
    def add(self):
        return self.val1+self.val2
    def sub(self):
        return self.val1-self.val2
    def mul(self):
        return self.val1*self.val2
    def div(self):
        return self.val1/self.val2
        
```

# The constructor
## constructor is a method with same name as a class

```python
cal = calculator(5,6)
```

```python
print(cal.mul())
```

# Inheritance:
## Parent and child relationship

```python
class human:
    legs = 2
    hands = 2
    arms = 2
    eyes = 2
    ears = 2

    def speak(self):
        print("Speaking ....")
    def walk(self):
        print("Walking !!!!")

class male(human):
    pass
    
```

```python
hafi = male()

print(hafi.legs)

hafi.speak()
```

# Polymorphism
1. Compile Time (Overloading)
2. Run Time (Overriding)

```python
class calculator:
    def __init__(self):
        pass
    def add(self,a,b):
        return a+b
    def add(self,a,b,c):
        return a+b+c
```

```python
class animal:
    legs = 0
    hands = 0
    arms = 0
    eyes = 0
    ears = 0
    def voice(self):
        print("Some voice...")

        
class cat(animal):
    def voice(self):
        print("Meow.Meow....")

class dog(animal):
    def voice(self):
        print("Barking.... Bhaow Bhaow ...")
```

```python
cat1 = cat()
dog1 = dog()
ani1 = animal()
```

```python
ani1.voice()
cat1.voice()
dog1.voice()
```

```python
class human:
    legs = 2
    hands = 2
    arms = 2
    eyes = 2
    ears = 2

    def speak(self):
        print("Speaking ....")
    def walk(self):
        print("Walking !!!!")
```

```python
class student(human):
    def __init__(self,id, name, age,marks = []):
        self.id = id
        self.name = name
        self.age = age
        self.marks = marks
    def reading(self):
        print(f'{self.name} is reading ....')
    def writing(self):
        print(f'{self.name} is writing ....')
```

```python
hafiullah = student('std_01','Hafiullah',20)
```

```python
mustafa = student('std_02','Mustafa',18)
```

```python
nofil = student('std_03','Nofil',18)
```

```python
mustafa.reading()
hafiullah.writing()
print(nofil.age)
```
